```python
from torch.utils.data import DataLoader, Dataset

class NERDataset(Dataset):
    def __init__(self, df):
        self.df = df
        self.tokens=['B-CORP', 'B-CW', 'B-GRP', 'B-LOC', 'B-PER', 'B-PROD', 'I-CORP', 'I-CW', 'I-GRP','I-LOC', 'I-PER', 'I-PROD', 'O']

    def __len__(self):
        return len(self.df) 

    def __getitem__(self, idx):
        sentence=self.df['Sentence'][idx]
        tokenized_sentence=tokenizer(sentence,truncation=True,padding='max_length',max_length=250)
        label=self.df['POS'][idx]
        #model_input = self.process_tokenized_sentence(tokenized_sentence)
        labels = self.process_label(sentence,label)
        labels = self.label_to_one_hot(labels)
        return {
            "input_ids": torch.tensor(tokenized_sentence['input_ids']).long(),
            "attention_masks": torch.tensor(tokenized_sentence['attention_mask']).long(),
            "labels": torch.tensor(labels)
        },torch.tensor(labels)
    def process_tokenized_sentence(self,tokenized_sentence):
        one_hot_list=[]
        for token in tokenized_sentence: 
            one_hot_list.append(nn.functional.one_hot(token, num_classes=105879))
        return torch.stack(one_hot_list)

    def label_stabilizer(self,sentence,label):
        sentence=sentence.split()
        label=label.split()
        ret=[]
        len_sentence=len(sentence)
        for i in range(len_sentence):
            token=tokenizer.tokenize(sentence[i],truncation=True)
            len_token=len(token)
            for _ in range(len_token):
                ret.append(label[i])
        return ret


    def process_label(self,sentence,label_list):
        #print(sentence,label_list)
        label_list=self.label_stabilizer(sentence,label_list)

        tokenized_label = [self.tokens.index(val) for val in label_list]


        tokenized_label = self.pad_list(tokenized_label)
        return torch.tensor(tokenized_label)
    def label_to_one_hot(self,label):
        one_hot_label_list=[]
        for token in label: 
            one_hot_label_list.append(nn.functional.one_hot(token, num_classes=13))
        y=torch.stack(one_hot_label_list)
        y = y.to(torch.float)
        return y

    def pad_list(self,my_list,length=250):
        padded_list = my_list + [12] * (length - len(my_list))
        return padded_list

    
train_dataset = NERDataset(train_df)
test_dataset = NERDataset(test_df)
```

```python
train_dataloader = DataLoader(train_dataset, batch_size=32, shuffle=True, pin_memory=True)
test_dataloader = DataLoader(test_dataset, batch_size=32, shuffle=True, pin_memory=True)
```

# F1 score function

```python
def compute_f1_score(y_true, y_pred):
    assert len(y_true) == len(y_pred)

    # Get the index of the maximum value for each example in y_pred
    y_pred = y_pred.argmax(dim=-1)
    y_true = y_true.argmax(dim=-1)
    
    #print(len(y_pred))
    #print(len(y_true))
    #print(y_pred)
    
    true_positive = 0
    false_positive = 0
    false_negative = 0
    for i in range(len(y_true)):
        for j in range(len(y_true[i])):
            if y_true[i][j] == y_pred[i][j]:
                true_positive += 1
            elif y_true[i][j] != y_pred[i][j] :
                false_positive += 1
            elif y_true[i][j] != y_pred[i][j]:
                false_negative += 1

    precision = true_positive / (true_positive + false_positive) if (true_positive + false_positive) != 0 else 0
    recall = true_positive / (true_positive + false_negative) if (true_positive + false_negative) != 0 else 0
    f1_score = 2 * (precision * recall) / (precision + recall) if (precision + recall) != 0 else 0

    return f1_score
```

# Training

```python
model.train()
model.to('cuda')
# Define the loss function and optimizer
#['B-CORP', 'B-CW', 'B-GRP', 'B-LOC', 'B-PER', 'B-PROD', 'I-CORP', 'I-CW', 'I-GRP','I-LOC', 'I-PER', 'I-PROD', 'O']
weights = [0.5592763664357198, 0.6736207695873899, 0.6041580041580042, 0.6180348787749894, 
           0.5575594781273984, 0.455771643663739, 0.5379489078119215, 0.5109001406469761, 
           0.3420433145009416, 1.0, 0.4639208173690933, 0.7398167006109979, 0.19567082683307333] #[ 1 / number of instances for each class]
class_weights = torch.FloatTensor(weights).cuda() 

criterion = torch.nn.CrossEntropyLoss(weight=class_weights)#)
optimizer = torch.optim.Adam(model.parameters(), lr=3e-5)




# Training loop
for epoch in range(10):
    f1_score_list=[]
    for data in tqdm(train_dataloader, total=int(len(train_dataset)/32),leave=True):
        # Encode the input
        #print(data[0])
        input_ids = data[0]["input_ids"]
        attention_mask = data[0]["attention_masks"]
        
        #input_ids = torch.tensor(input_ids).unsqueeze(0)
        #attention_mask = torch.tensor(attention_mask).unsqueeze(0)
        labels=data[0]["labels"]#.unsqueeze(0)
        # Forward pass
        output = model(input_ids.to('cuda'), attention_mask=attention_mask.to('cuda'))
        #print(output['logits'].shape)
        #print(labels.shape).transpose_(1, 2)
        loss = criterion(output['logits'].transpose(1, 2), labels.transpose(1, 2).to('cuda'))
        
        # Backward pass and optimization
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        

        #print(labels.shape)
        #print(output['logits'].shape)
        f1_score_list.append(compute_f1_score(labels.to('cuda'), output['logits'].to('cuda')))
        print(f"Epoch {epoch + 1}, Loss: {loss.item()},F1_score:{f1_score_list[-1]}",end='\r')
        
        
        
        del output
        del input_ids
        del attention_mask
        
        
        torch.cuda.empty_cache()
        gc.collect()
    # Print the loss for this epoch
    sum_of_score=0
    for i in f1_score_list:
        sum_of_score+=i
    sum_of_score/=len(f1_score_list)
    
    f1_score_test_list=[]
    for data in tqdm(test_dataloader, total=int(len(test_dataset)/32),leave=True):

        input_ids = data[0]["input_ids"]
        attention_mask = data[0]["attention_masks"]
        labels=data[0]["labels"]#.unsqueeze(0)
        output = model(input_ids.to('cuda'), attention_mask=attention_mask.to('cuda'))
        f1_score_test_list.append(compute_f1_score(labels.to('cuda'), output['logits'].to('cuda')))
        
        del output
        del input_ids
        del attention_mask
        
        
        torch.cuda.empty_cache()
        gc.collect()
        
    sum_of_test_score=0
    for i in f1_score_test_list:
        sum_of_test_score+=i
    sum_of_test_score/=len(f1_score_test_list)

    
    
    print(f'Epoch {epoch + 1}, Loss: {loss.item()}, Train F1:{sum_of_score},Test F1:{sum_of_test_score}')
```

```python
def predict(sentence):
    tokenized_sentence=tokenizer(sentence,truncation=True)
    input_ids=torch.tensor(tokenized_sentence['input_ids']).long()
    attention_masks= torch.tensor(tokenized_sentence['attention_mask']).long()
    
    input_ids=torch.unsqueeze(input_ids, 0)
    attention_masks=torch.unsqueeze(attention_masks,0)
    
    predicted_val=model(input_ids.to('cuda'), attention_mask=attention_masks.to('cuda'))
    return torch.nn.functional.softmax(torch.squeeze(predicted_val['logits']),dim=-1)
```
